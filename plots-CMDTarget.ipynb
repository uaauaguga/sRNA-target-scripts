````python
energy_shuffling = pd.read_csv("/BioII/lulab_b/jinyunfan/projects/sRNATarget-revised/evaluate-shuffling.txt",sep="\t",index_col=0)
score = "loc"
correlations = []
rmses = []
ground_truth = energy_shuffling[energy_shuffling["#"] == 100][score].iloc[:500]
for i in sorted(np.unique(energy_shuffling["#"].values[:-1])):
    x = energy_shuffling[energy_shuffling["#"] == i][score].iloc[:500]
    diff = ground_truth-x
    RMSE = np.sqrt((diff*diff).mean())
    #print(ground_truth.shape,x.shape)
    r = pearsonr(ground_truth,x)[0]
    correlations.append(r)
    rmses.append(RMSE)
    #print(i,r,MSE)
times = np.unique(energy_shuffling["#"].values[:-1])
fig, ax = plt.subplots(figsize=(3,2))
ax.bar(x=np.array(times[1:-1]).astype(str),height=rmses[1:-1],width=0.5)
ax.plot([-0.5,9.5],[rmses[0],rmses[0]],zorder=1,color="black")
ax.set_xlim([-0.5,9.5])  
ax.set_ylabel("RMSE")
ax.set_xlabel("# of shuffling")
plt.savefig("figures/bg-performances/gumbel-loc.pdf",bbox_inches="tight")
````

````python
def plot_hfq_weight(sRNA_ids,genome_id, validated, hfq_weight,notation=None,outdir="test"):
    intaRNA_total_recalls = np.zeros(500)
    sRNA_hfq_total_recalls = np.zeros(500) 
    pair_hfq_total_recalls = np.zeros(500) 
    pair_total_recalls = np.zeros(500) 
    N = 0
    stats = defaultdict(int)
    for sRNA_id in sRNA_ids:
        target_path =  f"output/Enterobacteriaceae/{genome_id}/combined.hfq.200.100.32/{sRNA_id}.txt"
        recalls, n = get_performance(sRNA_id,genome_id,target_path,field='sRNA zscore',validated=validated,id_extractor = lambda x:x.split("-")[0])
        if n is None:
            continue
        comparisons = []
        if n >= 5:
            use = True
            comparisons.append((recalls[199]/n,"IntaRNA"))
        
        
        intaRNA_total_recalls += recalls
        recalls, n = get_performance(sRNA_id,genome_id,target_path,field='sRNA zscore',hfq_weight=hfq_weight,id_extractor = lambda x:x.split("-")[0],
                                     validated=validated,hfq_field = "hfq Z")
        
        if use:
            comparisons.append((recalls[199]/n,f"Hfq lambda={hfq_weight},unadjusted"))
        sRNA_hfq_total_recalls += recalls
        recalls, n = get_performance(sRNA_id,genome_id,target_path,field='pair zscore',hfq_weight=hfq_weight,id_extractor = lambda x:x.split("-")[0],
                                     validated=validated,hfq_field = "hfq Z")
        if use:
            comparisons.append((recalls[199]/n,f"Hfq lambda={hfq_weight},adjusted"))
        pair_hfq_total_recalls += recalls
        recalls, n = get_performance(sRNA_id,genome_id,target_path,field='pair zscore',hfq_weight=0,id_extractor = lambda x:x.split("-")[0],
                                     validated=validated,hfq_field = "hfq Z")
        if use:
            comparisons.append((recalls[199]/n,"IntaRNA,adjusted"))
        pair_total_recalls += recalls
        
        rs = []
        for r, m in sorted(comparisons,key=lambda x:-x[0]):
            rs.append(r)
        if np.std(rs) > 0:
            rank = 0
            for _, m in sorted(comparisons,key=lambda x:-x[0]):
                stats[(m,rank)] += 1
                rank += 1
        N += n
    fig, ax = plt.subplots(figsize=(2.5,2))
    ax.plot(intaRNA_total_recalls/N,label="IntaRNA") 
    ax.plot(pair_total_recalls/N,label=f"InatRNA,adjusted")
    ax.plot(sRNA_hfq_total_recalls/N,label=f"Hfq $\lambda$={hfq_weight},unadjusted")
    ax.plot(pair_hfq_total_recalls/N,label=f"Hfq $\lambda$={hfq_weight},adjusted")    
    ax.set_xlabel("# predictions")
    ax.set_ylabel("Recalls")
    #ax.set_title(genome_id + " " + ",".join(sRNA_ids))
    plt.legend(frameon=False,loc=(1.05,0.5))
    records = []
    for m, rank in stats:
        c = stats[(m,rank)]
        records.append((m,rank,c))
    table = pd.DataFrame.from_records(records)
    table.columns = ["method","rank","count"]
    return table
    #if notation is None:
    #    plt.savefig(f"figures/hfg-weights.leader/{genome_id}-{validated}-{hfq_weight}.pdf",bbox_inches="tight")
    #else:
    #    plt.savefig(f"figures/hfg-weights.leader/{notation}.{genome_id}-{validated}-{hfq_weight}.pdf",bbox_inches="tight")
````

````python
def get_params_performance(sRNA_ids,asm_id,params,validated=True,outdir="output/Enterobacteriaceae",excluded_protein_ids=None):
    total_recalls = np.zeros(500)
    N = 0
    for sRNA_id in sRNA_ids:    
        predicted_targets_path = f"{outdir}/{asm_id}/combined.final/{params}/{sRNA_id}.txt"
        #if not os.path.exists(predicted_targets_path):
        #    print(predicted_targets_path)
            
        recall, n = get_performance(sRNA_id,asm_id,predicted_targets_path,
                            min_homolog=0,field='denoised score',hfq_weight=0.0,hfq_field = "hfq bg zscore",
                            id_extractor = lambda x:x.split(":")[1].split("-")[0],validated=validated,excluded_protein_ids=excluded_protein_ids)
        if recall is None:
            continue
        total_recalls += recall
        N += n
        #print(sRNA_id,n)
    #print(N)
    return total_recalls/N
````
